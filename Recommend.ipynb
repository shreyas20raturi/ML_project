**Popularity Based Recommender**

```python
num_rating_df=rating_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df)
popular_df
```

**Collaborating Filtering Based Recommender System**

```python
x = rating_name.groupby("User-ID").count()["Book-Rating"] > 200
User_Rec=x[x].index
User_Rec
```

```python
filtered_rating=rating_name[rating_name["User-ID"].isin(User_Rec)]
```

```python
y = filtered_rating.groupby('Book-Title').count()["Book-Rating"]>=50
famous_book = y[y].index
```

```python
famous_book
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns="User-ID",values="Book-Rating")
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
print(similarity_score)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
  #index fetch
  index = np.where(pt.index == book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse = True)[1:6]

  for i in similar_items:
    print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```
